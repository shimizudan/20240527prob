```julia

#f(n)は10本中あたりが1本のくじをn回引いた時の当たりくじを引く割合
function f(n=10^6)
    t = 0
    for i = 1:n
        if rand(1:10) == 7 # 7が当たり 
            t += 1
        end
    end
    t/n
end

#100万回を10セット
for i = 1:10
    @show f(10^6)
end
```

```julia
x = []
k = 0
n = 10^4
for i = 1:n
    k += rand(1:10) == 7
    append!(x,k/i)
end

using Plots
plot(x,label = false)
```
